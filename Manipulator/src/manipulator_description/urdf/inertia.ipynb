```python
import numpy as np

def cuboid_inertia(w, h, d, m):

    Ixx = (1/12) * m * (h**2 + d**2)
    Iyy = (1/12) * m * (w**2 + d**2)
    Izz = (1/12) * m * (w**2 + h**2)

    I = np.array([
        [Ixx, 0.0, 0.0],
        [0.0, Iyy, 0.0],
        [0.0, 0.0, Izz]
    ])

    return I


def cylinder_inertia(r, h, m):

    Ixx = (1/12) * m * (3*(r**2) + h**2)
    Iyy = Ixx
    Izz = (1/2) * m * (r**2)

    I = np.array([
        [Ixx, 0.0, 0.0],
        [0.0, Iyy, 0.0],
        [0.0, 0.0, Izz]
    ])

    return I
```

```python
def shift_inertia(I_com, m, r_com):
    """
    Shift inertia tensor from COM to reference point P=(0,0,0)
    using the parallel axis theorem.
    
    r_com = (x, y, z) position of COM relative to P
    """
    r = np.array(r_com)
    d2 = np.dot(r, r)          # |d|^2
    I3 = np.identity(3)

    return I_com + m * (d2 * I3 - np.outer(r, r))
```

```python
w1 = 0.045   # meters
h1 = 0.022
d1 = 0.044
m1 = 0.075   # kg

I1 = cuboid_inertia(w1, h1, d1, m1)
print(I1)
```
